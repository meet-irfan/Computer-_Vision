```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Load the image
image = cv2.imread('/content/irfan.jpeg')
if image is None:
    raise FileNotFoundError("Image not found. Please check the path.")

rows, cols = image.shape[:2]

# 1. Translation matrix: shift right 50px, down 30px
M_translate = np.float32([[1, 0, 50], [0, 1, 30]])
translated = cv2.warpAffine(image, M_translate, (cols, rows))

# 2. Rotation matrix: rotate 45 degrees around the center
M_rotate = cv2.getRotationMatrix2D((cols / 2, rows / 2), 45, 1)
rotated = cv2.warpAffine(image, M_rotate, (cols, rows))

# Display results using matplotlib
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original")

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(translated, cv2.COLOR_BGR2RGB))
plt.title("Translated")

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB))
plt.title("Rotated")

plt.tight_layout()
plt.show()
```
